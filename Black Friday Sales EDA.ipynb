```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
bfs_train=pd.read_csv(r"E:\EDA PROJECT\DATASETS FOR EDA\DATASET FOR BLACK FRIDAY\train.csv")
bfs_train.head()
```

```python
bfs_test=pd.read_csv(r"E:\EDA PROJECT\DATASETS FOR EDA\DATASET FOR BLACK FRIDAY\test.csv")
bfs_test.head()
```

```python
bfs_train.shape
```

```python
bfs_test.shape
```

```python
bfs=bfs_train._append(bfs_test)
bfs.head()
```

```python
bfs.shape
```

```python
bfs.info()
```

```python
bfs.drop(['User_ID'],axis=1,inplace=True)
bfs.head()
```

```python

bfs['Gender']=bfs['Gender'].map({'F':0,'M':1})
```

```python
bfs.head()
```

```python

bfs['Age'].unique()
```

```python

from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
bfs['Age']= label_encoder.fit_transform(bfs['Age']) 
bfs['Age'].unique() 
```

```python
bfs.head()
```

```python

bfs_city=pd.get_dummies(bfs['City_Category'],drop_first=True,dtype=int)
```

```python
bfs_city.head()
```

```python
bfs=pd.concat([bfs,bfs_city],axis=1)
```

```python
bfs.head()
```

```python
bfs.drop('City_Category',axis=1,inplace=True)
```

```python
bfs.head()
```

```python
# Handling missing values
bfs.isna().sum()
```

```python

bfs['Product_Category_2'].unique()
```

```python
bfs['Product_Category_2'].value_counts()
```

```python

bfs['Product_Category_2']=bfs['Product_Category_2'].fillna(bfs['Product_Category_2'].mode()[0])
```

```python
bfs['Product_Category_2'].isnull().sum()
```

```python
bfs['Product_Category_3'].value_counts()
```

```python

bfs['Product_Category_3']=bfs['Product_Category_3'].fillna(bfs['Product_Category_3'].mode()[0])
```

```python
bfs['Product_Category_3'].isna().sum()
```

```python

bfs['Stay_In_Current_City_Years'].unique()
```

```python
bfs['Stay_In_Current_City_Years']=bfs['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
bfs['Stay_In_Current_City_Years']
```

```python
bfs.head()
```

```python

bfs['Stay_In_Current_City_Years']=bfs['Stay_In_Current_City_Years'].astype(int)
bfs.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set a clean style
sns.set(style="whitegrid")

# Use a boxplot instead of barplot to show distribution of Purchase amounts across Age groups and Gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='Purchase', hue='Gender', data=bfs)

plt.title('Purchase Distribution by Age and Gender')
plt.xlabel('Age Group')
plt.ylabel('Purchase Amount')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```
